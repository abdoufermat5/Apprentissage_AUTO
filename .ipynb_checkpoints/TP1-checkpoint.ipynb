```python
import numpy as np
import pandas as pd
import random
import time
import functools

import matplotlib.pyplot as plt
```

```python
def generer_jeux_donnees(taille):
    
    X = 2 * np.random.rand(taille, 1)  
    y = 4 + 3 * X + np.random.randn(taille, 1)
    return X, y
```

```python
X, y = generer_jeux_donnees(100)
```

```python
X.shape
```

```python
y.shape
```

```python
# Calculer θ à l’aide de l’équation normale.
def time_decorator(func):
    @functools.wraps(func)
    def wrapper_time(*args, **kwargs):
        debut = time.perf_counter()
        val = func(*args, **kwargs)
        fin = time.perf_counter()
        exec_time = fin - debut
        print("|{:>20}:{:>10.5f} secondes".format("temps d'execution", exec_time))
        return val
    return wrapper_time

@time_decorator
def get_theta(X, y):
    """fonction qui retourne theta"""
    theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    return theta
```

```python
get_theta(X, y)
```

```python
# En ajoutant un biais (vecteur de 1)
X_b = np.c_[np.ones((100, 1)), X]
```

```python
def affiche_params(X, y):
    print('-'*41)
    print("{:-^41}".format("paramètres du modele"))
    print('-'*41)
    num_par = X.shape[1]
    global theta
    theta = get_theta(X, y)
    for i in range(num_par):
        print("|{:>20}:{:>10.2f}".format("theta_%d"%i, theta[i][0]))
    print('-'*41)
    print('-'*41)
```

```python
affiche_params(X_b, y)
```

```python
# interpretation:
"""
On peut voir que les parametres du modele ont été bien approximés
"""
```

```python

def predire(X_test, theta):
    y_pred = X_test.dot(theta)
    return y_pred
```

```python
X_test = np.array([[0], [2]])
X_test_biais = np.c_[np.ones((2, 1)), X_test]
y_pred = predire(X_test_biais, theta)
```

```python
def affiche_predict(y_pred, X_test):
    print('-'*41,"\n{:-^41}".format("predictions"), '-'*41, '\n')
    print()
    print("|{:>10}|{:>10}|{:>10}".format('X_test', 'y_pred', 'y_reel'))
    print('-'*41)
    y = 4 + 3 * X_test + np.random.randn(X_test.shape[0], 1)
    i = 0
    for (y_p, y_r) in zip(y_pred, y):
        print("|{:>10.3f}|{:>10.3f}|{:>10.3f}".format(X_test[i][0], y_p[0], y_r[0]))
        i += 1
```

```python
affiche_predict(y_pred, X_test)
```

```python
# representation
def plot_graphe_ML(titre="Representation graphique du modele lineaire"):
    fig, ax = plt.subplots(figsize=(10, 8))
    f_1 = plt.plot(X, y, 'y^', markersize=8)
    f_2 = plt.plot(X_test, y_pred, color="g", Marker="o",linestyle="--", linewidth=4)
    plt.axis([-0.05, 2.05, 0, 13])
    plt.xlabel("axe X")
    plt.xlabel("axe y")
    plt.title(titre)
    ax.legend(['nuage', 'droite'], loc = "upper left", prop={'size': 15})
```

```python
# donnees non-lineaires
def generer_data_non_lineaire(n_samples):
    """renvoi un jeu de données non lineaire"""
    de_linearize = lambda X: 3 * np.cos(1.5 * np.pi * X) + 2 * np.cos( 5 * np.pi * X )
    X = np.sort(np.random.rand(n_samples)) * 4
    y = de_linearize(X) + np.random.randn(n_samples) * 0.1 + 4
    return X, y
```

```python
# y = 0.5 * x ^ 2 + x + 2.0 + bruit_gaussien
def polynome_y(taille, X, add_noise = False):
    if add_noise:
        noise = np.random.rand(taille, 1)
        y = 0.5 *( X ** 2) + X + 2 + noise
        return y
    y = 0.5 * (X ** 2) + X + 2
    return y
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python

poly_feature = PolynomialFeatures(degree=2)
X_p = poly_feature.fit_transform(X)
```

```python
# entrainons un model lineaire sur le dataset
model = LinearRegression()
model.fit(X_p, y)
```

```python
y_pred = model.predict(X_p)
```

```python
print('RootMeanSquaredError: {:>5.3f}\nr_square: {:>5.3f}'.format(
                                            np.sqrt(mean_squared_error(y, y_pred)), 
                                            r2_score(y, y_pred)))
plt.plot(X, y, 'bo', label='points')
plt.plot(sorted(X[:, 0]), y_pred[np.argsort(X[:, 0])], 'r', label='Regression Polynomiale')
plt.legend()
plt.show()
```

```python
def predictions(X, y, degre):
    pol_F = PolynomialFeatures(degree=degre)
    X_pol = pol_F.fit_transform(X)
    model = LinearRegression().fit(X_pol, y)
    return model.predict(X_pol)
    
```

```python
degrees = [1, 2, 8, 16, 30]
cs = ['r', 'g', 'c', 'y', 'k']

plt.figure(figsize=(10, 6))
plt.plot(X, y, 'b+', label='points')

for degree, c in zip(degrees, cs):
    # Make predictions for degree
    preds = predictions(X, y, degree)
    # Plot
    plt.plot(sorted(X[:, 0]), preds[np.argsort(X[:, 0])], c, label='Degree: {}'.format(degree))
    # Print Metrics
    print('Degrés: {}\tRMSE: {:.3f}'
          '\tR2: {:.3f}'.format(degree, np.sqrt(mean_squared_error(y, preds)), r2_score(y, preds)))

plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# fonction 
def visualiser_ca(model, X, y, split=0.2, metric=mean_squared_error):
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=split)
    train_errors, test_errors = [], []
    for train_size in range(1, len(x_train)):
        model.fit(x_train[:train_size], y_train[:train_size])
        preds_train = model.predict(x_train[:train_size])
        
        y_pred = model.predict(x_test)
        
        train_errors.append(metric(y_train[:train_size], preds_train))
        test_errors.append(metric(y_test, y_pred))
        
    # representation
    plt.plot(np.sqrt(train_errors), 'r--*', label='Erreurs entrainement')
    plt.plot(np.sqrt(test_errors), 'b--+', label='Erreurs test')
    plt.ylabel('RMSE')
    plt.xlabel('Training set size')
```

```python
plt.figure(figsize=(10, 6))
visualiser_ca(model, X, y)
plt.legend()
plt.show()
```

```python
# Essayons avec un model lineaire degree=1
from sklearn.pipeline import Pipeline
"""
Pipeline permet de combiner PolynomialFeatures et un model de regression
quelconque
"""
model = Pipeline([
    ('pol_F', PolynomialFeatures(degree=1, include_bias=False)),
    ('linearModel', LinearRegression())
])
plt.figure(figsize=(10, 6))
visualiser_ca(model, X, y)
plt.legend()
plt.show()
```

```python
# IMPLEMENTATION
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet


def get_model(model, model_name, X, y, alpha, ratio=False):
    if alpha == 0:
        if ratio:
            model = Pipeline([
                ('poly_feats', PolynomialFeatures(degree=3)),
                (model_name, model(alpha=alpha, l1_ratio=0.5))
            ])
        else:
            model = Pipeline([
                ('poly_feats', PolynomialFeatures(degree=3)),
                (model_name, LinearRegression())
            ])
    elif alpha != 0:
        if ratio:
            model = Pipeline([
                ('poly_feats', PolynomialFeatures(degree=3)),
                (model_name, model(alpha=alpha, l1_ratio=0.5))
            ])
            
        else:
            model = Pipeline([
                ('poly_feats', PolynomialFeatures(degree=3)),
                (model_name, model(alpha=alpha))
            ])
        
    model.fit(X, y)
    return model.predict(X)
```

```python
alphas = [0, 0.7, 5]
cs = ['r', 'g', 'y']

plt.figure(figsize=(10, 6))
plt.plot(X, y, 'b+', label='points')
print('Model Lasso')
for alpha, c in zip(alphas, cs):
    preds = get_model(Lasso, 'lasso', X, y, alpha)
    plt.plot(sorted(X[:, 0]), preds[np.argsort(X[:, 0])], c, label='Alpha: {}'.format(alpha))

plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Ridge
alphas = [0, 20, 200]
cs = ['r', 'g', 'b']

plt.figure(figsize=(10, 6))
plt.plot(X, y, 'b+', label='Datapoints')
print('Ridge model')
for alpha, c in zip(alphas, cs):
    preds = get_model(Ridge, 'ridge', X, y, alpha)
    # Plot
    plt.plot(sorted(X[:, 0]), preds[np.argsort(X[:, 0])], c, label='Alpha: {}'.format(alpha))

plt.legend()
plt.show()
```
